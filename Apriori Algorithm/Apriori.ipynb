```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from apyori import apriori
from scipy.io import arff
```

```python
%%time
records = []
for i in range(0,df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(df.shape[1])])
```

```python
records
```

```python
%%time
#Building the apriori model
association_rules = apriori(records, min_support = 0.15, min_confidence = 0.1)
type(association_rules)
association_results = list(association_rules)
```

```python
#Getting the number of rules
print(len(association_results))
```
